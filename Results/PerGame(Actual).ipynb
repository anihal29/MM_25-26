```python
import pandas as pd
```

##### Final Matchup Dataset

```python
# Final CSV based on matchups that actually happened
actual_preds = pd.read_csv("2026 March Madness Scoring - Ayaan.csv")[4:67]
actual_preds_columns = ['T1', 'T2',	'Pred_Likelihood', 'Pred_Win', 'Actual', 'Brier']
actual_preds = actual_preds[actual_preds_columns]
actual_preds['Pred_Win'] = pd.to_numeric(actual_preds['Pred_Win'], errors='coerce')
actual_preds['Actual'] = pd.to_numeric(actual_preds['Actual'], errors='coerce')
actual_preds = actual_preds.reset_index()
actual_preds = actual_preds.drop(columns = 'index')
pd.set_option('display.max_rows', None)
actual_preds
```

##### Final Brier Score and Total Number of Games Predicted Correctly

```python
brier_score = round(float(actual_preds.Brier.mean()), 4)
games_correctly_predicted = int((actual_preds.Pred_Win == actual_preds.Actual).astype(int).sum())

print(f"Brier Score of {brier_score}\n{games_correctly_predicted}/63 games correctly predicted ({round(games_correctly_predicted/63 * 100, 2)}% accuracy)")
```
